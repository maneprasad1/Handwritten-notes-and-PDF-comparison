```python
def get_marathi_ocr(filename):
    # Read the image
    image = cv2.imread(filename)

    # Convert the image to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Perform OCR
    marathi_text = pytesseract.image_to_string(image_rgb, lang='mar', config=f"--tessdata-dir {os.environ.get('TESSDATA_PREFIX')}")

    return marathi_text
```

```python
def process_images(input_folder, output_folder):

    for filename in os.listdir(input_folder):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            # Full path to the input image
            image_path = os.path.join(input_folder, filename)

            marathi_ocr_output = get_marathi_ocr(image_path)

            # Create the output folder if it doesn't exist
            os.makedirs(output_folder, exist_ok=True)

            output_file_path = os.path.join(output_folder, f'{os.path.splitext(filename)[0]}.txt')
            with open(output_file_path, 'w', encoding='utf-8') as file:
                file.write(marathi_ocr_output)

            print(f"Extracted Marathi text saved to: {output_file_path}")

# Specify the input and output folders
input_folder = r'/content/drive/MyDrive/OCR/output_1'
output_folder = r'/content/drive/MyDrive/OCR/ocr_output1'

# Process images and perform OCR
process_images(input_folder, output_folder)
```
